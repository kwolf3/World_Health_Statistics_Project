```python
import pandas as pd
import pathlib
```

Select out the data I want from the BMI dataset.

```python
datasets_path = pathlib.Path.cwd()/'Datasets'
```

```python
BMI = pd.read_csv(datasets_path/'NCD_BMI_30A.csv')
BMI
```

```python
BMI_2016_messy = BMI[['Unnamed: 0', '2016']]
BMI_2016_messy
```

```python
BMI_2016 = BMI_2016_messy.drop(BMI_2016_messy.index[0:3])
BMI_2016
```

```python
BMI_2016.columns = ['Location', 'BMI30Plus']
BMI_2016
```

## Combine Data

```python
test = pd.read_csv('test_small_cleaner_with_tobacco.csv')
test
```

```python
test_BMI = pd.merge(test, BMI_2016, on = "Location", how = "left")
```

```python
test_BMI
```

```python
test_BMI.to_csv('test_small_messy_with_tobacco_BMI.csv', index = False)
```
