```python
import pandas as pd
import pathlib
```

Select out the data I want from the BMI dataset.

```python
datasets_path = pathlib.Path.cwd()/'Datasets'
```

```python
BMI = pd.read_csv(datasets_path/'NCD_BMI_30A.csv')
BMI
```

```python
BMI_2015_messy = BMI[['Unnamed: 0', '2015']]
BMI_2015_messy
```

```python
BMI_2015 = BMI_2015_messy.drop(BMI_2015_messy.index[0:3])
BMI_2015
```

```python
BMI_2015.columns = ['Location', 'BMI30Plus']
BMI_2015
```

# Combine Data

```python
training = pd.read_csv('training_small_cleaner_with_tobacco.csv')
training
```

```python
training_BMI = pd.merge(training, BMI_2015, on = "Location", how = "left")
```

```python
training_BMI
```

```python
training_BMI.to_csv('training_small_cleaner_with_tobacco_BMI.csv', index = False)
```
