```python
import pandas as pd
import pathlib
```

```python
new_data_path = pathlib.Path.cwd()/'New Data'
```

```python
TechFuel_2016 = pd.read_csv(new_data_path/'TechFuel_2016.csv')
TechFuel_2016
```
