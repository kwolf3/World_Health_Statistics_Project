```python
import pandas as pd
import pathlib
```

```python
new_data_path = pathlib.Path.cwd()/'New Data'
```

```python
TechFuel_2015 = pd.read_csv(new_data_path/'TechFuel_2015.csv')
TechFuel_2015
```

```python
TechFuel_2015.columns = ['Location', 'Indicator', 'Period', 'CleanTechFuel']
```

```python
TechFuel_2015
```

```python
TechFuel_2015 = TechFuel_2015.drop(['Period', 'Indicator'], 1)
TechFuel_2015
```

# Combine Datasets

```python
training = pd.read_csv('training_small_cleaner_with_tobacco_BMI.csv')
training
```

```python
training_TechFuel = pd.merge(training, TechFuel_2015, on = "Location", how = "left")
```

```python
training_TechFuel
```

```python
training_TechFuel[training_TechFuel.isnull().any(axis=1)].count()
```

```python
training_TechFuel.to_csv('training_small_cleaner_with_tobacco_BMI_TechFuel.csv', index = False)
```
